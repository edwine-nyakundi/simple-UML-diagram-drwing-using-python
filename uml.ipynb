```python
import graphviz

uml_diagram = graphviz.Digraph('UML_Diagram', format='png')

# Warehouse class
uml_diagram.node('Warehouse', shape='rectangle', label='Warehouse\n- totes: List<Tote>')
uml_diagram.node('Warehouse_loadProducts', label='+ loadProducts()')
uml_diagram.node('Warehouse_fulfillOrders', label='+ fulfillOrders()')
uml_diagram.node('Warehouse_mergeTotes', label='+ mergeTotes()')
uml_diagram.node('Warehouse_displayDetails', label='+ displayDetails()')

uml_diagram.edge('Warehouse', 'Warehouse_loadProducts')
uml_diagram.edge('Warehouse', 'Warehouse_fulfillOrders')
uml_diagram.edge('Warehouse', 'Warehouse_mergeTotes')
uml_diagram.edge('Warehouse', 'Warehouse_displayDetails')

# Tote class
uml_diagram.node('Tote', shape='rectangle', label='Tote\n- products: List<Product>')
uml_diagram.node('Tote_isFull', label='+ isFull()')
uml_diagram.node('Tote_isEmpty', label='+ isEmpty()')
uml_diagram.node('Tote_addProduct', label='+ addProduct()')

uml_diagram.edge('Tote', 'Tote_isFull')
uml_diagram.edge('Tote', 'Tote_isEmpty')
uml_diagram.edge('Tote', 'Tote_addProduct')

# Product class
uml_diagram.node('Product', shape='rectangle', label='Product\n- upc: String')
uml_diagram.node('Product_getUpc', label='+ getUpc()')

uml_diagram.edge('Product', 'Product_getUpc')

# Render the UML diagram
uml_diagram.render(filename='UML_Diagram', format='png', cleanup=True)

# Display the UML diagram
from IPython.display import Image
Image('UML_Diagram.png')
```
